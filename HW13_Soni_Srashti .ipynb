### 2. Do the manual pivoting of the data by
  1. Grouping the data by 'Description and District'
  2. Selecting 'Neighborhood'
  3. Aggregating the data using count
  4. `unstack()` the table

```python
neighbor_crime.groupby(['Description', 'District'])['Neighborhood'].aggregate('count').unstack()
```

### 3. Using Pandas `pivot_table()` method, produce the same outcome as above (using manual groupby)

```python
neighbor_crime_pivot_table = neighbor_crime.pivot_table(index='Description', columns='District', values='Neighborhood', aggfunc='count')
```

### 4. Using 'CrimeDate' column, create a DatetimeIndex
Hint: 
- You will use one of the methods described in 'Pandas Time Series Data Structures' section. 
- You will have to specify the values of the `Series` as the input. See: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.values.html

```python
dates = pd.to_datetime(neighbor_crime['CrimeDate'], format="%Y-%m-%d")
date_index = pd.DatetimeIndex(dates.values)

#setting the index of your DataFrame to the new DatetimeIndex
dates=neighbor_crime.set_index(date_index)
dates.drop('CrimeDate',axis=1,inplace=True)

dates.sort_index(inplace=True)
dates.index
```
